A cloth manufacturing company is interested to know about the segment or attributes causes high sale.

Approach - A decision tree can be built with target variable Sale (we will first convert it in categorical variable) & all other variable will be independent in the analysis.

```python
import numpy as np
import pandas as pd
df=pd.read_csv("Company_Data.csv")
df
```

```python
median_sales = df["Sales"].median()
```

```python
df["Sales"]=["High" if value > median_sales
             else "Low" for value in df["Sales"]]
df
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
Columns_to_encoder=["Sales","ShelveLoc","Urban","US"]
for column in Columns_to_encoder:
  df[column]=LE.fit_transform(df[column])
df
```

```python
X = df.iloc[:,1:]
X
```

```python
Y = df["Sales"]
Y
```

```python
df.corr()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y)
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
DTC.fit(X_train,Y_train)
```

```python
Y_pred_train = DTC.predict(X_train)
Y_pred_test = DTC.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
ac1=accuracy_score(Y_train,Y_pred_train)
ac2=accuracy_score(Y_test,Y_pred_test)
print("Accuracy score Training Data:",ac1.round(3))
print("Accuracy score Test Data:",ac1.round(3))
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC=RandomForestClassifier(n_estimators=150,
                           max_samples=0.6,
                           max_features=0.7)
RFC.fit(X_train,Y_train)
```

```python
Y_pred_train = RFC.predict(X_train)
Y_pred_test = RFC.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
ac1=accuracy_score(Y_train,Y_pred_train)
ac2=accuracy_score(Y_test,Y_pred_test)
print("Random Forest for Training Data:",ac1.round(3))
print("Random Forest for Test Data:",ac1.round(3))
```
